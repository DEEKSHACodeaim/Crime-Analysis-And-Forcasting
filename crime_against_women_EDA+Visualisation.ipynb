```python
cor=['STATE/UT','DISTRICT','Year','Rape','Kidnapping_and_Abduction','Dowry_Deaths','Assault_to_outrage_her_modesty','Insult_to_modesty_of_Women','Cruelty_by_Husband/Relatives','Importation_of_Girls']
```

```python
fig = plt.figure(figsize=(10,10))
gs = fig.add_gridspec(1,1)
gs.update(wspace=0.3, hspace=0.15)
ax0 = fig.add_subplot(gs[0,0])

color_palette = ["#5833ff","#da8829"]
mask = np.triu(np.ones_like(cor))
ax0.text(1.5,-0.1,"Correlation Matrix",fontsize=22, fontweight='bold', fontfamily='serif', color="#000000")
df_corr = df[cor].corr().transpose()
sns.heatmap(df_corr,fmt=".1f",annot=True,cmap='YlGnBu')
plt.show()
```
